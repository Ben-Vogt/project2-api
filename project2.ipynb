## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed
page = requests.get("https://en.wikipedia.org/wiki/Batman").text
```

```python
with open("batman_page.html", "w", encoding='utf-8') as file:
    file.write(page)
```

```python
soup = BeautifulSoup(page)
```

```python
print(soup.get_text())
```

```python
#download an image with beautifulsoup and save it in this repository
images = soup.find_all("img")
images[4]["src"]

```

```python
img = requests.get("https:"+images[4]["src"])
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags
def transformLink(link):
    return link if "http" in link else "http://" + link

count = 0
texts = []
for link in scripts["link"]:
    html = "lol"
    try:
        html = requests.get(transformLink(link)).text
    except Exception as e:
        count += 1
    soup = BeautifulSoup(html)
    texts.append(soup.get_text()[:1000])
print(count)
```

```python
import tmdbsimple as tmdb
```

I used a seperate file not tracked by the github repo to house the API key I obtained, to avoid posting it on github. To run this code you will likely need your own valid API key. 

```python
from dotenv import load_dotenv
import os

# Load the api key from the .env file in the same directory as this notebook
load_dotenv()
tmdb.API_KEY = os.getenv("TMDB_API_KEY")

#Set a request timeout of 5 seconds, because I'm impatient.
tmdb.REQUESTS_TIMEOUT = 5
```

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 
mov = tmdb.Movies()
now_playing = mov.now_playing()
movies = []
for page in range(1, now_playing["total_pages"] + 1):
    data = mov.now_playing(page=page)
    movies.extend(data["results"])
    
print(len(movies))
```

I guess I kind of just included all the metadata fields... I don't really have much interest in movies anyways.

```python
movie_df = pd.DataFrame(movies)
movie_df
```

```python
#download the movie posters for 10 of these movies and save them to this repository
base_url = tmdb.Configuration().info()["images"]["base_url"]
for i in range(0, 10):
    img = requests.get(base_url+movie_df["poster_path"][i])
    with open(movie_df["title"][i]+".jpg", "wb") as file:
        file.write(img.content)
```
